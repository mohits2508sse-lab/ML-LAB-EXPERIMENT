```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

X, y = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=1.0,
    random_state=42
)

gmm = GaussianMixture(
    n_components=3,
    random_state=42
)

gmm.fit(X)

labels = gmm.predict(X)

print("Means of Clusters:")
print(gmm.means_)

print("\nCovariances of Clusters:")
print(gmm.covariances_)

print("\nWeights of Clusters:")
print(gmm.weights_)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(
    gmm.means_[:,0],
    gmm.means_[:,1],
    marker='X',
    s=200
)

plt.title("Expectation-Maximization using Gaussian Mixture Model")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
